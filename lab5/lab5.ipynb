__Работа с пропущенными значениями (если они есть): удаление строк с пропущенными значениями, заполнение пропущенных значений средним значением по колонке. Если пропущенных значений нет — намеренно их "генерируете", прибивая какие-то куски данных в np.nan__

```python
df.loc[0, 'Age'] = np.nan
df.loc[1, 'Age'] = np.nan
df.loc[2, 'Age'] = np.nan
df.loc[1, 'Genre'] = np.nan
df
```

```python
df['Age'] = df['Age'].fillna(
    np.ceil(df['Age'].mean()))
df
```

```python
df = df.dropna()
df
```

__Cоздание нового поля вычисленного на основе значений других полей:__

*через выражение на базе имеющихся колонок*

```python
df.loc[:,'AgeID'] = (df.loc[:, 'Age'] + df.loc[:, 'CustomerID']) / 2
df
```

*через DataFrame.apply*

```python
def caculate(row):
    return (row['Age'] + row['CustomerID']) / 2

df.loc[:, 'AgeID'] = df.apply(caculate, axis=1)
df
```

*через Series.apply*

```python
def s_calculate(value):
    return value < 35

df.loc[:, 'Bool'] = df.loc[:, 'Age'].apply(s_calculate)
df
```
